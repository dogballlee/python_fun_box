```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
sns.set(font='SimHei')
```

```python
# path = 'C:\\Users\\lilinyang\\Desktop\\mpl_test.csv'
# df_test = pd.read_csv(path, encoding = 'UTF-8')
# data1 =df_test.iloc[:,2:]
# df_test.head(5)
path1 = 'C:\\Users\\lilinyang\\Desktop\\mpltest.xlsx'
df_test1 = pd.read_excel(path1).fillna(0)
course_list = list(set(df_test1.iloc[:,1]))
# df_test1.head(5)
# gp = df_test1.groupby(['course_name','日期'])
# print(gp.sum())
pv = pd.pivot_table(df_test1,index=[u'日期'],columns=[u'course_name'],values=[u'订单数'],aggfunc=[np.sum],fill_value=0).reset_index()
pv.columns = [s0 + '_'+ s2 for (s0,s1,s2) in pv.columns.tolist()]
```

```python
course_list = pv.columns.tolist()
print(course_list)
for col in course_list[1:]:
    plt.subplot(7,2,course_list.index(col))
    g = sns.lineplot(x='日期_',y=pv[col],data=pv)
    plt.xlabel('日期', fontsize=10)
    plt.ylabel(col, fontsize=7)
plt.show()
```
